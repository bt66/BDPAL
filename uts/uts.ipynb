<h2>2.	Baca data dari FullData.csv, kemudian pecah setidaknya menjadi 2 RDD (players, clubs) (10, SCPMK 1534104)</h2>

membaca dataset menggunakan pyspark lalu menampilkan isi dataset nya

```python
# read dataset csv using pyspark
dataset = spark.read.csv('/content/drive/MyDrive/kuliah/Smester 5/big data lanjut/uts/FullData.csv', header=True, inferSchema=True)
dataset.show()
```

memecah dataset menjadi 2 RDD, saya mengelompokkan data dengan melakukan seleksi kolom kemudian mengkonversi hasil seleksi kolom menjadi rdd

```python
# select column from dataset 
players = dataset.select('Name','Nationality','National_Position','Age')
clubs = dataset.select('Club','Club_Position','Club_Kit','Club_Joining')
# create rdd from selected column
playersRDD = players.rdd
clubsRDD = clubs.rdd
# ensure data type is rdd
print('type playersRDD :', type(playersRDD))
print('type clubsRDD :', type(clubsRDD))
```

<h1>3.	Ubah RDD di atas menjadi PySpark Dataframe (5, SCPMK 1532205)</h1>

convert rdd ke Pyspark dataframe dapat di lakukan dengan memanggil atribut toDF() dari rdd yang telah di buat sebelumnya

```python
# convert rdd to pyspark Dataframe
playersPysparkDataframe = playersRDD.toDF()
clubsPysparkDataframe = clubsRDD.toDF()
# ensure data type is pyspark dataframe
print('type playersPysparkDataframe :', type(playersPysparkDataframe))
print('type playersPysparkDataframe :', type(clubsPysparkDataframe))
```

<h1> 5.	Lakukan agregasi untuk mendapatkan nilai: (25, SCPMK 1534104) </h1>

import Function dari pyspark sql dan mengaliaskan sebagai F

```python
# import function from pyspark.sql library and make alias F
from pyspark.sql import functions as F
```

a.	Rata-rata Rating berdasarkan Club_Position

melakukan grouping berdasarkan Club_Position kemudian melakukan fungsi agregasi avg untuk mencari rata rata Rating dan mengaliaskan menjadi avgRating

```python
# grouping by Club_Position and aggregate to count average Rating
dataset.groupBy("Club_Position").agg(
    F.avg("Rating").alias("avgRating")
).show()
```

b.	Maximum Rating berdasarkan National_Position

melakukan grouping berdasarkan Club_Position kemudian melakukan fungsi agregasi MAX untuk mencari nilai maksimum Rating dan mengaliaskan menjadi avgRating 

```python
# group by ClubPosition and aggregate to find max rating 
dataset.groupBy("Club_Position").agg(
    F.max("Rating").alias("maxRating")
).show()
```

<h1>6.	Lakukan setidaknya 3 operasi transform RDD pada data yang kalian punya, contoh reduceByKey, sortByKey, flatMap, etc. (30, SCPMK 1532206)</h1>

Melakukan transform RDD berupa mapping data menggunakan map() untuk mendapatkan key dan value yang kita mau

saya akan melakukan mapping dengan tujuan menetapkan umur sebagai key dan nama sebagai value

```python
# set age as key and Name as value
ageAndNameRDD = playersRDD.map(lambda x: (x[3],x[0]))
# collect data from rdd
ageAndNameRDD.collect()
```

Melakukan transform RDD berupa sorting umur secara ascending menggunakan sortByKey()

pertama saya melakukan mapping terhadap playersRDD sehingga menghasilkan umur sebagai key dan nama sebagai value nya

setelah di mapping kemudian saya sort secara ascending sehingga hasilnya seperti di bawah ini : 

```python
# sorting ascending by key (Age) 
sortedageAndNameRDD = ageAndNameRDD.sortByKey(ascending=True)
# collect sorted RDD
sortedageAndNameRDD.collect()
```

transform rdd berupa filter dengan memanggil atribut filter() dari rdd


dalam dataset ini saya melakukan filter dengan tujuan mendapatkan Nationality dengan value France

```python
# filter data Nationality which value is France
playersItaly = playersRDD.filter(lambda x : (x[1]) == 'France')
# collect filtered data
playersItaly.collect()
```
