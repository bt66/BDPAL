```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
print(spark)
print(spark.version)
```

identify by key and value
pada list inted ke [0] value nya 1,2 sehingga 1 merupakan key dan 2 sebagai value

jika key nya sama tetapi value nya berbeda maka akan di jumlahkan

```python
Rdd = sc.parallelize([(1,2),(3,4),(3,6),(4,5)])
Rdd_Reduced = Rdd.reduceByKey(lambda x, y: x+y)
for num in Rdd_Reduced.collect(): 
    print("Key {} has {} Counts".format(num[0], num[1]))
```

```python
#ambil data
Rdd_Reduced.collect()
```

```python
# mengurutkan data kemudian collect
Rdd_Reduced_Sort = Rdd_Reduced.sortByKey(ascending=True)
for num in Rdd_Reduced_Sort.collect():
  print("Key {} has {} Counts".format(num[0], num[1]))
```

```python
# mengambil count berdasarkan key
total = Rdd.countByKey()

print("The type of total is", type(total))

for k, v in total.items(): 
    print("key", k, "has", v, "counts")
```

```python
# import file yang sudah di download
file_path = '/content/drive/MyDrive/kuliah/Smester 5/big data lanjut/p7/Complete_Shakespeare.txt'
```

```python
# menghitung kata di dataset
baseRDD = sc.textFile(file_path)

splitRDD = baseRDD.flatMap(lambda x: x.split())

print("Total number of words in splitRDD:", splitRDD.count())
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
print(stop_words)
```

```python
# Ubah ke lowercase dan abaikan stop_words
splitRDD_no_stop = splitRDD.filter(lambda x: x.lower() not in stop_words)

# Buat tuple dari setiap kata
splitRDD_no_stop_words = splitRDD_no_stop.map(lambda w: (w, 1))

resultRDD = splitRDD_no_stop_words.reduceByKey(lambda x, y: x + y)
```

```python
#collect only 10 index from rdd
for word in resultRDD.take(10):
	print(word) 

# Tukar key dan value
resultRDD_swap = resultRDD.map(lambda x: (x[1], x[0]))

resultRDD_swap_sort = resultRDD_swap.sortByKey(ascending=False)
```

```python
for word in resultRDD_swap_sort.take(10):
	print("{},{}". format(word[1], word[0]))
```
